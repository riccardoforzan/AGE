```python
import os
import json
```

```python
json_datasets_file = "ACORDAR/Data/datasets.json"
```

```python
f = open(json_datasets_file)
data = json.load(f, strict=False)
f.close()
```

```python
datasets = sorted(data["datasets"], key=lambda d: int(d['dataset_id']))
```

```python
link_dataset = dict()

for entry in datasets:
    dataset_id = entry["dataset_id"]
    for download_link in entry["download"]:
        if download_link not in link_dataset.keys():
            link_dataset[download_link] = list()

        l = link_dataset[download_link]
        l.append(dataset_id)
```

```python
def compare_datasets(datasets: list, ids: list):

    # key = urls , value = dataset IDs that downloads the same URLs
    d = dict()

    for id in ids:
        entry = next(filter(lambda e: e["dataset_id"] == id, datasets))

        # use as key a sorted tuple containing the downloaded URLs
        key = str(sorted(entry["download"]))
        if key not in d:
            d[key] = list()
    
        d[key].append(id)

    d_rep = json.dumps(d, indent=4)
    
    if len(d.items()) == 1:
        print(f"DUPLICATED: {d_rep}")
    else:
        print(f"PROBABLY DUPLICATED: {d_rep}")
    

    
```

```python
for url, datasets_list in link_dataset.items():
    if len(datasets_list) > 1:
        compare_datasets(datasets, datasets_list)
        
```
